```python
# Numerical Operations
import math
import numpy as np
from sklearn.manifold import TSNE
import seaborn as sns
# Reading/Writing Data
import pandas as pd
import os
import csv

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
```

```python
train_x = pd.read_csv('../ML_storage/train_binary_x.csv')
train_y = pd.read_csv('../ML_storage/train_binary_y.csv')
test_x = pd.read_csv('../ML_storage/test_binary_x.csv')
test_y = pd.read_csv('../ML_storage/test_binary_y.csv')
```

```python
tsne = TSNE(n_components=2, verbose=1, random_state=123)
z = tsne.fit_transform(test_x) 
```

```python
df = pd.DataFrame()
df["y"] = test_y
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]

sns.scatterplot(x="comp-1", y="comp-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 3),
                data=df).set(title="Iris data T-SNE projection") 
```

```python
tsne = TSNE(n_components=3, verbose=1, random_state=123)
z = tsne.fit_transform(test_x) 
```

```python
fig = plt.figure(figsize=(6,6))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

cmap = ListedColormap(sns.color_palette("husl", 256).as_hex())

trans_data = z

sc = ax.scatter(trans_data[:,0], trans_data[:,1], trans_data[:,2], s=40, c=test_y, marker='o', alpha=1)

plt.legend(*sc.legend_elements(), bbox_to_anchor=(1.05, 1), loc=2)

```
